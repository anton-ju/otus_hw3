##### ●	Разбейте весь временной промежуток в данных на 10 частей и для каждой найдите самое популярное имя каждого пола. Для каждого найденного имени визуализируйте его динамику за все время

```python

#slice_ind = 
sliced = [years[x:x+13] for x in [years.index(_) for _ in years[::13]]]
sliced[0].append(1880)
print(len(sliced))
print(sliced)
```

```python
for s in sliced:
    dfsliced1 = df[np.isin(df['Year'].values, s)]
    mname = dfsliced1[dfsliced1.loc[:,'Sex']=='M'].groupby(['Name'])['Number'].sum().nlargest().index[0]
    fname = dfsliced1[dfsliced1.loc[:,'Sex']=='F'].groupby(['Name'])['Number'].sum().nlargest().index[0] 

    df[(df.Name==mname) & (df.Sex=='M')].groupby('Year').sum().plot(title='{0} most popular since {1} until {2}'.format(mname, s[0], s[12]))
    df[(df.Name==fname) & (df.Sex=='F')].groupby('Year').sum().plot(title='{0} most popular since {1} until {2}'.format(fname, s[0], s[12]))
```

###### ●	Для каждого года рассчитайте сколько имен покрывают 50% людей, визуализируйте (мы увидим разнообразие имен за каждый год)

```python
cover = []
for y in df.Year.unique():
    
    dfsorted= df.loc[df.loc[:,'Year']==y,['Name', 'Number']].sort_values(by=['Number'], ascending=False)
    half = dfsorted.Number.sum()/2
    cumsum = dfsorted.Number.cumsum()
    cover.append(len(cumsum[cumsum<half]))
    
```

```python
cover = pd.DataFrame(data=cover, columns=['Number'], index=df.Year.unique())
cover.plot()
```
